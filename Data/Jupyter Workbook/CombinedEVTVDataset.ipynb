```python
# Import the dependencies
import pandas as pd
```

```python
# load csv file
EV = pd.read_csv('../EVData.csv')
EV.head()
```

```python
# load csv file
TV = pd.read_csv('../TVData.csv')
TV.head()
```

```python
combinedDF = pd.concat([EV, TV])
combinedDF
```

```python
# Export file as a CSV, without the Pandas index, but with the header
combinedDF.to_csv("../combinedData.csv", index=False, header=True)
```
