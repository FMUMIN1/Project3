```python
# Import the dependencies
import pandas as pd
```

```python
# load csv file
dataDF = pd.read_csv('../Raw Data/CO2 Emissions_Canada.csv')
dataDF.head()
```

```python
# Check unique value of vehicle class
Type = dataDF["Vehicle Class"].unique()
Type
```

```python
# Rename column
dataDF = dataDF.rename(columns={"Vehicle Class": "Type"})
dataDF.head()
```

```python
# Rename elements 
dataDF['Type'].replace({'SUV - SMALL':'SUV', 'MINICOMPACT':'COMPACT', 'SUBCOMPACT':'COMPACT', 'STATION WAGON - SMALL':'SUV', 'SUV - STANDARD':'SUV', 'VAN - CARGO':'VAN', 'VAN - PASSENGER': 'VAN', 'PICKUP TRUCK - STANDARD': 'TRUCK', 'MINIVAN':'VAN','STATION WAGON - MID-SIZE':'SUV','PICKUP TRUCK - SMALL':'TRUCK'}, inplace=True)
dataDF.head()
```

```python
Type = dataDF["Type"].unique()
Type
```

```python
# Rename Columns and convert (L/100 km) to (MPG)
dataDF = dataDF.rename(columns={"Fuel Consumption City (L/100 km)": "Fuel Consumption City (MPG)",
                               "Fuel Consumption Hwy (L/100 km)": "Fuel Consumption Hwy (MPG)"})
dataDF.head()
```

```python
# Convert l/100km to mpg 
dataDF["Fuel Consumption City (MPG)"] = 235.215 / dataDF["Fuel Consumption City (MPG)"]
dataDF["Fuel Consumption Hwy (MPG)"] = 235.215 / dataDF["Fuel Consumption Hwy (MPG)"]
dataDF.head()
```

```python
# Round numbers to 2 decimals
dataDF.round({"Fuel Consumption City (MPG)":2, "Fuel Consumption Hwy (MPG)":2}).head()
```

```python
# Rename Columns and convert CO2 Emission (g/km) to (g/miles)
dataDF = dataDF.rename(columns={"CO2 Emissions(g/km)": "CO2 Emissions(g/m)"})
dataDF.head()
```

```python
# convert CO2 Emission (g/km) to (g/miles)
dataDF["CO2 Emissions(g/m)"] =  dataDF["CO2 Emissions(g/m)"] * 1.60934
dataDF.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
dataDF.to_csv("../TradCarData.csv", index=False, header=True)
```
